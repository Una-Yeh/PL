```python
import pandas as pd

data = pd.read_csv('常見濫用管制藥品資料集.csv')
data.head()
```

```python
# 如果需要中文的詞頻統計，可以使用jieba這個中文分詞庫
# pip install jieba，為安裝jieba庫的命令

import pandas as pd
import jieba
from jieba import analyse
import jieba.analyse

jieba.analyse.set_stop_words('stopwords.txt')

data['seg'] = ''
index = 0
for documents in data['說明']:
    seg_list = jieba.lcut(documents, cut_all=False)
    seg_list = jieba.analyse.extract_tags(documents)
    data['seg'].iloc[index] = seg_list
    index += 1
data["seg"]
```

```python
seg_list = data['seg'].tolist()
```

```python
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer

words = []

for text in seg_list:
    for word in text:
        if word not in words:
            words.append(word)
            
# 空的矩陣
matrix = np.zeros((len(words), len(words)))

# 填充矩陣
for text in seg_list:
    for i in range(len(text)):
        word1 = text[i] # 當前所在的索引
        for j in range(len(text)): # 當前索引的下一個詞語
            word2 = text[j]
            index1 = words.index(word1) # 獲取詞語1在詞語列表的索引
            index2 = words.index(word2) # 獲取詞語2在詞語列表的索引
            matrix[index1, index2] += 1 # 增加共現記數

matrix
```

```python
data = pd.DataFrame(matrix, index=words, columns=words)
data
```

```python
import plotly.express as px

fig = px.imshow(data)

fig.update_layout(
    title='共現性 Heatmap',
    xaxis=dict(title='管制藥品濫用常見詞語'),
    yaxis=dict(title='管制藥品濫用常見詞語'),
)

fig.show()
```

```python
# 找出共現性較高的詞語

# 共現次數域值，可根據數據調整
threshold = 3 

for i in range(len(words)):
    if matrix[i, i] > threshold:
        print(words[i], matrix[i, i])
```
