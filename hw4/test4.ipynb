```python
import pandas as pd

data = pd.read_csv('常見濫用管制藥品資料集.csv')
data.head()
```

```python
import pandas as pd
import jieba
from jieba import analyse
import jieba.analyse

jieba.analyse.set_stop_words('stopwords.txt')
data['seg'] = ''

index = 0
for documents in data['類別']:
    seg_list = jieba.lcut(documents, cut_all=False)
    seg_list = jieba.analyse.extract_tags(documents)
    # seg_list = [word for word in seg_list if word not in stopwords]
    data['seg'].iloc[index] = seg_list
    index += 1
    
data["seg"]
```

```python
seg_list = data['seg'].tolist()
print(seg_list)
```

```python
from collections import Counter

word_count = Counter([word for doc in seg_list for word in doc])
word_count
```

```python
count = pd.DataFrame.from_dict(word_count, orient='index', columns=['count'])
count = count.sort_values(by=['count'], ascending=False) # 按照數值大小降序排列
count
```

```python
words = []
for doc in data['類別']:
    for word in doc.split():
        words.append(word)
unique_words = set(words)
unique_words
```

```python
from collections import Counter

count = Counter(words)
print(count)
```

```python
words = list(data["類別"])
show_key = pd.DataFrame(Counter(words).keys())
show_values = pd.DataFrame(Counter(words).values())
df = [show_key, show_values]
result = pd.concat(df, axis=1)
result.columns = ['category', 'count']
result
```
