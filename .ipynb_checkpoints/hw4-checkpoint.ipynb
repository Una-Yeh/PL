```python
import pandas as pd

data = pd.read_csv('常見濫用管制藥品資料集.csv')
data.head()
```

```python
# 如果需要中文的詞頻統計，可以使用jieba這個中文分詞庫
# pip install jieba，為安裝jieba庫的命令

import pandas as pd
import jieba
from jieba import analyse
import jieba.analyse

# 停用詞列表(需另外下載)

jieba.analyse.set_stop_words('stopwords.txt')

# 將 index (追蹤目前正在處理的資料的索引位置)初始化為 0
# 使用精確模式斷詞 (cut_all=False)，會選擇機率最大的詞作為斷詞結果
# 使用 extract_tags函式提取關鍵詞，會根據關鍵詞的權重進行排序
# 移除停用詞
# 將列表儲存在「seg」欄位
# 將變數 index的值循環加 1，即可處理下一個元素

data['seg'] = ''
index = 0
for documents in data['說明']:
    seg_list = jieba.lcut(documents, cut_all=False)
    seg_list = jieba.analyse.extract_tags(documents)
    seg_list = [word for word in seg_list if word not in stopwords]
    data['seg'].iloc[index] = seg_list
    index += 1
data["seg"]
```

```python
seg_list = data['seg'].tolist()
print(seg_list)
```

```python
# 計算各詞語出現次數

from collections import Counter

word_count = Counter([word for doc in seg_list for word in doc])
word_count
```

```python
count = pd.DataFrame.from_dict(word_count, orient='index', columns=['count'])
count.index.name = 'word'
count = count.reset_index()
count.head(10)
```

```python
# pip install wordcloud，為安裝文字雲程式的命令

from wordcloud import WordCloud
import matplotlib.pyplot as plt

# 創建一個 WordCloud物件
# 生成文字雲

wc = WordCloud(background_color="white", width=800, height=400, font_path='msjh.ttc') # msjh.ttc 是微軟正黑體字型
wc.generate_from_frequencies(word_count)

# 顯示圖片

plt.figure(figsize=(12,6))
plt.imshow(wc, interpolation='bilinear') # interpolation='bilinear'表示雙線性插值，使影像更加平滑自然
plt.axis("off")
plt.show()
```

```python
# 生成不同圖案的文字雲

from PIL import Image
import numpy as np

# 使用自定義的圖形作為文字雲形狀
mask = np.array(Image.open("9518.png"))
wc = WordCloud(background_color = "white", font_path= 'msjh.ttc', mask = mask)
wc.generate_from_frequencies(word_count)

plt.figure(figsize=(12,6))
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
import plotly.graph_objects as go

top_words = dict(word_count.most_common(30))  # 出現次數前30高的詞語

fig = go.Figure([go.Bar(x=list(top_words.keys()), y=list(top_words.values()))])
fig.update_layout(
    title='常見濫用管制藥品出現次數前 30 高的詞語',
    xaxis_title='藥品說明常見詞語',
    yaxis_title='出現頻率',
    font=dict(size=14),
    width=1000,
    height=500
)
fig.show(renderer='iframe')
```
