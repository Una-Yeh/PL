```python
import pandas as pd

data = pd.read_csv('adult.data.csv' , header = None)
data.head()
```

```python
# 將標頭名稱設置為表格的第一行
data.columns = ['age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital-status',
              'occupation', 'relationship', 'race', 'sex', 'capital-gain', 'capital-loss',
              'hours-per-week', 'native-country', 'income']
data.head()
```

```python
data.count()
```

```python
# 1. How many men and women (sex feature) are represented in this dataset?
# 計算每個性別的出現次數
counts = data['sex'].value_counts()

# 建立一個新的 DataFrame，並將人數顯示在表格中
gender = pd.DataFrame({'Gender': counts.index, 'Count': counts.values})
gender
```
